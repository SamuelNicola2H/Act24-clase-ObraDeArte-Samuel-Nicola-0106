```python
class ObraDeArte:
    def __init__(self, titulo, artista, anio, tecnica):
        # Atributos: Datos referentes a la obra
        self.titulo = titulo
        self.artista = artista
        self.anio_creacion = anio
        self.tecnica = tecnica

    # Funciones: Acciones que realiza la obra
    def exhibir(self):
        print(f"\n> Acción: La obra '{self.titulo}' ha sido colgada en la sala de exhibición.")

    def restaurar(self):
        print(f"> Acción: Iniciando restauración de la técnica {self.tecnica} para la obra de {self.artista}.")

    def tasvalor(self):
        print(f"> Acción: El perito está calculando el valor comercial de la pieza creada en {self.anio_creacion}.")

def crear_obra(numero):
    print(f"\n--- DATOS DE LA OBRA DE ARTE {numero} ---")
    tit = input("Título de la obra: ")
    art = input("Nombre del Artista: ")
    ani = input("Año de creación: ")
    tec = input("Técnica utilizada: ")

    # Crear objeto (Instancia)
    objeto = ObraDeArte(tit, art, ani, tec)

    # Ejecutar funciones (Acciones)
    objeto.exhibir()
    objeto.restaurar()
    objeto.tasvalor()
    return objeto

# --- Ejecución del Programa ---

# Se crean los 2 objetos solicitados
ObraArte1 = crear_obra(1)
ObraArte2 = crear_obra(2)

print("\n--- PROCESO DE GALERÍA FINALIZADO ---")
print("Programa realizado por Samuel Nicola NL = 40")
```
